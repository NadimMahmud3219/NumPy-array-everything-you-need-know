1. What is NumPy?

NumPy (Numerical Python) is a Python library used for:

Fast mathematical operations

Efficient storage of large datasets

Linear algebra, matrices, statistics

Foundation for libraries like Pandas, SciPy, Scikit-learn, TensorFlow, PyTorch

Its core feature is the ndarray (N-dimensional array).



2. NumPy Array (ndarray) Overview

A NumPy array is:

Homogeneous (all elements same data type)

Fixed-size

Fast (implemented in C)

Supports vectorization (operations on entire arrays without loops)
